BankAccount Class to store account details (name, account number, balance) in a data structure.

```python
import random
from datetime import datetime

class BankAccount:
    """
    Bank account class, storing customer details, balance, account number, and transaction history
    """
    def __init__(self, name, initial_deposit):
        """
        Initialize bank account

        Args:
            name (str): Account holder's name
            initial_deposit (float): starting deposit amount

        Raises:
            ValueError: If initial deposit is invalid (negative or non-float)
        """
        # Store account holder's name
        self.name = name
        
        # Generate 7-digit account number
        self.account_number = self.generate_account_number()

        # Create empty transaction list
        self.transactions = []

        try:
            initial_deposit = float(initial_deposit)

             # Raise ValueError if initial deposit is negative
            if initial_deposit < 0:
                raise ValueError
            else:
                # Set initial balance
                self.balance = initial_deposit

                if initial_deposit > 0: 
                     # Append initial deposit to transaction list
                    self.transactions.append({
                        "date": datetime.now().strftime("%m/%d/%Y %H:%M:%S"), # 00/00/0000 00:00:00
                        "type": "deposit",
                        "amount": initial_deposit
                    })

        except ValueError:
            print("Invalid initial deposit. Balance set to $0.00.")
            self.balance = 0.0


    def generate_account_number(self):
        """
        Generate a random 7-digit account number
        
        Returns:
            int: 7-digit account number
        """
        account_number = random.randint(1000000, 9999999)
        return account_number


    def deposit(self, amount):
        """
        Deposits money into account and updates balance
        
        Args:
            amount (float): amount to deposit

        Raises:
            ValueError: If amount is invalid (negative or non-float)
        """
        try:
            amount = float(amount)

            # Deposit amount must be positive
            if amount <= 0:
                print("Deposit amount must be greater than $0.00.")
                return
            
            # Add deposit amount to balance
            self.balance += amount

            # Append transaction
            self.transactions.append({
                "date": datetime.now().strftime("%m/%d/%Y %H:%M:%S"),
                "type": "deposit",
                "amount": amount
            })
        
            print(f"Deposited ${amount:.2f} successfully. New balance: ${self.balance:.2f}")

        except ValueError:
            print("Invalid deposit amount. Please enter a numeric value.")


    def withdraw(self, amount):
        """
        Withdraws money from account and updates balance

        Args:
            amount (float): amount to withdraw

        Raises:
            ValueError: If amount is invalid (negative or non-float)
        """
        try:
            amount = float(amount)

            # Withdrawl amount must be positive
            if amount <= 0:
                print("Withdrawal amount must be greater than $0.00.")
                return
            
            # Check for sufficient funds
            if amount > self.balance:
                print("Insufficient funds for withdrawal.")
                return
            
            # Subtract withdrawal amount from balance
            self.balance -= amount

            # Append transaction
            self.transactions.append({
                "date": datetime.now().strftime("%m/%d/%Y %H:%M:%S"),
                "type": "withdrawal",
                "amount": amount
            })
            
            print(f"Withdrew ${amount:.2f} successfully. New balance: ${self.balance:.2f}")

        except ValueError:
            print("Invalid withdrawal amount. Please enter a numeric value.")


    def display_account_info(self):
        """
        Displays account information

        Shows account holder's name, account number, and current balance
        """
        print("\nAccount Information")
        print("-" * 40)
        print(f"Name: {self.name}")
        print(f"Account Number: {self.account_number}")
        print(f"Balance: ${self.balance:.2f}")
        print("-" * 40)
        print()


    def display_withdrawals_above(self, amount):
        """
        Displays withdrawals above a certain amount

        Args:
            amount (float): amount to compare and display
        """
        try:
            amount = float(amount)

            # Withdrawal amount must be positive
            if amount <= 0:
                print("Negative withdrawal amount. Displaying withdrawals above $0.00.")
                amount = 0

            print(f"\nWithdrawals above ${amount:.2f}:")
            print("-" * 40)
            found = False

            # Iterate through each entry in transactions
            for transaction in self.transactions:
                # Display withdrawal transactions that are greate than amount
                if transaction["type"] == "withdrawal" and transaction["amount"] > amount:
                    print(
                        f"{transaction['date']} - "
                        f"{transaction['type'].title()}: "
                        f"${transaction['amount']:.2f}"
                    )
                    found = True

            print("-" * 40)

            if not found:
                print(f"There are no withdrawals made above ${amount:.2f}.")

        except ValueError:
            print("Invalid withdrawal amount. Please enter a numeric value.")


    def display_transactions_by_index(self, start_index, end_index):
        """
        Displays transactions through specified range

        Args:
            start_index (int): start index, inclusive
            end_index (int): end index, inclusive
        """
        try:
            start_index = int(start_index)
            end_index = int(end_index)
            total_transactions = len(self.transactions)

            # Check if starting index is valid and less then end index
            if start_index < 1 or start_index > end_index:
                print("Invalid transaction index range.")
                return
            
            # Check if end index is valid
            if end_index > total_transactions:
                print(f"Invalid range. This account only has {total_transactions} transactions.")
                return
            
            # Get transactions in range, start_index - 1 because list is zero-based
            transactions_range = self.transactions[start_index - 1:end_index]

            # Check if no transactions in range
            if not transactions_range:
                print("No transactions found in the specificed range.")
                return
            
            print(f"\nTransactions from {start_index} to {end_index}:")
            print("-" * 40)

            # Displah each transaction in range
            for transaction in transactions_range:
                print(
                    f"{transaction['date']} - "
                    f"{transaction['type'].title()}: "
                    f"${transaction['amount']:.2f}"
                )

            print("-" * 40)

        except ValueError:
            print("Invalid transaction index. Please enter whole numbers.")
```

Holding Class for all bank accounts.

```python
class Bank:
    """
    Bank class containing multiple bank accounts
    """
    def __init__(self, name):
        """
        Initialize bank

        Args:
            name (str): bank name
        """
        self.name = name

        # Dictionary allows for efficient lookup
        self.accounts = {}


    def add_account(self, account):
        """
        Adds BankAccount object to bank

        Args:
            account (BankAccount): the account object to add

        """
        self.accounts[account.account_number] = account
        print(f"Account for {account.name} added successfully. Account number: {account.account_number}")

        
    def find_account(self, account_number):
        """
        Find account by account number

        Args:
            account_number (int): account number to search for   

        Returns:
            BankAccount: the account object if found
        """
        try:
            account_number = int(account_number)

            # Check if account exists
            if account_number in self.accounts:
                account = self.accounts[account_number]
                return account
            
            print("Account not found.")
            return
            
        # If a non 7-digit number is entered, this error message will be prompted as well
        except ValueError:
            print("Invalid account number. Please enter a 7-digit number.")
```

Following is the sample code to test the major banking functionalities specified in the problem statement. Please change and test your code with different inputs. 

```python
# Create new bank
big_bank = Bank("Big Bank")

# Create an account
first_account = BankAccount("Bruce Lee", 1000)

# Store account in the bank
big_bank.add_account(first_account)

# Perform transactions
first_account.deposit(500)
first_account.deposit(400)
first_account.deposit(300)
first_account.withdraw(200)
first_account.withdraw(300)
first_account.withdraw(400)

# Display account information
first_account.display_account_info()
first_account.display_withdrawals_above(200)
first_account.display_transactions_by_index(2, 3)

# Display account information by account number
account_number = 1951267
big_bank.find_account(account_number)
```
